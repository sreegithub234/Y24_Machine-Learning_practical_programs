```python
import numpy as np
import pandas as pd
from sklearn.datasets import load_iris
from sklearn.model_selection import KFold, cross_val_score
from sklearn.linear_model import LogisticRegression

# -----------------------
# 1. Load Dataset
# -----------------------
iris = load_iris()
X = iris.data
y = iris.target

# -----------------------
# 2. Define Model
# -----------------------
model = LogisticRegression(max_iter=200)

# -----------------------
# 3. Define K-Fold
# -----------------------
k = 5
kfold = KFold(n_splits=k, shuffle=True, random_state=42)

# -----------------------
# 4. Apply Cross Validation
# -----------------------
cv_scores = cross_val_score(model, X, y, cv=kfold, scoring="accuracy")

# -----------------------
# 5. Show Results
# -----------------------
results = {
    f"Fold {i+1}": score for i, score in enumerate(cv_scores)
}
results["Mean Accuracy"] = np.mean(cv_scores)
results["Std Deviation"] = np.std(cv_scores)

df_results = pd.DataFrame([results])
print(" Logistic Regression Performance (5-Fold CV):\n")
print(df_results.round(3))
```

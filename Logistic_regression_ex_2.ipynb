```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LogisticRegression

# ------------------------
# 1. Dataset (Marks vs Selection)
# ------------------------
X = np.array([[30], [40], [50], [55], [65], [70], [80], [90]])
y = np.array([0, 0, 0, 1, 1, 1, 1, 1])  # 0=Not Selected, 1=Selected

# ------------------------
# 2. Train Logistic Regression Model
# ------------------------
model = LogisticRegression()
model.fit(X, y)

# ------------------------
# 3. Prediction for new student (x=60)
# ------------------------
x_new = np.array([[60]])
predicted_class = model.predict(x_new)[0]
predicted_prob = model.predict_proba(x_new)[0, 1]

print("Predicted Probability of Selection:", predicted_prob)
print("Resultant Class:", predicted_class)

# ------------------------
# 4. Plot Logistic Regression Curve
# ------------------------
X_test = np.linspace(20, 100, 200).reshape(-1, 1)   # range of marks
y_prob = model.predict_proba(X_test)[:, 1]          # probability of class=1

plt.figure(figsize=(8,6))

# Plot sigmoid curve
plt.plot(X_test, y_prob, color="blue", linewidth=2, label="Logistic Curve")

# Scatter actual data points
plt.scatter(X, y, color="red", marker="o", label="Training Data")

# Mark the prediction point (x=60)
plt.scatter(x_new, predicted_class, color="green", s=100, marker="x", label="Prediction (x=60)")

# Formatting
plt.axhline(0.5, color="gray", linestyle="--", label="Decision Boundary (0.5)")
plt.xlabel("Entrance Marks")
plt.ylabel("Probability of Selection")
plt.title("Logistic Regression: Student Selection Prediction")
plt.legend()
plt.grid(True)
plt.show()
```

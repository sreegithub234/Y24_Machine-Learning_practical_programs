```python
import matplotlib.pyplot as plt
import numpy as np

# Positive class points
positive_points = np.array([[4, 1], [4, -1], [6, 0]])

# Negative class points
negative_points = np.array([[1, 0], [0, 1], [0, -1]])

# Plot the points
plt.scatter(positive_points[:, 0], positive_points[:, 1], color='blue', label='Positive class')
plt.scatter(negative_points[:, 0], negative_points[:, 1], color='red', label='Negative class')

# Hyperplane: x + y - 2 = 0 → y = -x + 2
x_vals = np.linspace(-3, 7, 100)
y_vals = -x_vals + 2
plt.plot(x_vals, y_vals, 'k-', label='Hyperplane: x + y - 2 = 0')

# Mark points with coordinates
for (x, y) in np.vstack([positive_points, negative_points]):
    plt.text(x + 0.1, y + 0.1, f"({x},{y})", fontsize=9)

# Axis formatting
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)
plt.xlim(-3, 7)
plt.ylim(-3, 4)
plt.xlabel("X")
plt.ylabel("Y")
plt.title("SVM Example: Hyperplane x + y - 2 = 0")
plt.legend()
plt.grid(True, linestyle="--", alpha=0.6)
plt.show()
```

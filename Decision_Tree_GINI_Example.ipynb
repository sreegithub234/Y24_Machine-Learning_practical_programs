build a Decision Tree using the Gini Index on the dataset (Weather, Parents, Money â†’ Decision).

```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder
from sklearn.tree import DecisionTreeClassifier, plot_tree
import matplotlib.pyplot as plt
```

```python
# Step 1: Create dataset
data = {
    'Weather': ['Sunny', 'Sunny', 'Windy', 'Rainy', 'Rainy', 'Rainy', 'Windy', 'Windy', 'Windy', 'Sunny'],
    'Parents': ['Yes', 'No', 'Yes', 'Yes', 'No', 'Yes', 'No', 'No', 'Yes', 'No'],
    'Money':   ['Rich', 'Rich', 'Rich', 'Poor', 'Rich', 'Poor', 'Poor', 'Rich', 'Rich', 'Rich'],
    'Decision': ['Cinema', 'Tennis', 'Cinema', 'Cinema', 'Stay In', 'Cinema', 'Cinema', 'Shopping', 'Cinema', 'Tennis']
}

df = pd.DataFrame(data)
```

```python
# Step 2: Encode categorical variables
le_dict = {}
for column in df.columns:
    le = LabelEncoder()
    df[column] = le.fit_transform(df[column])
    le_dict[column] = le  # save encoders for decoding later
```

```python
# Step 3: Split features and target
X = df.drop('Decision', axis=1)
y = df['Decision']
```

```python
# Step 4: Train Decision Tree with Gini index
clf = DecisionTreeClassifier(criterion="gini", random_state=0)
clf.fit(X, y)
```

```python
# Step 5: Plot decision tree
plt.figure(figsize=(12,8))
plot_tree(clf, feature_names=X.columns, class_names=le_dict['Decision'].classes_, filled=True, rounded=True)
plt.show()
```

# üß©* Support Vector Machine (SVM)*

SVM is a supervised machine learning algorithm mainly used for classification. Imagine you have two groups of points on a graph:

üî¥ Class A ‚Üí Red dots

üîµ Class B ‚Üí Blue dots

The goal of SVM is to draw the best line (or boundary) to separate these two groups. It chooses the line that leaves the widest gap (margin) between the two classes. The few points that lie closest to this line are called support vectors‚Äîthey are the most important points because they decide where the line will be placed.

**‚ú® Types of SVM**

Linear SVM ‚Üí Used when data can be separated by a straight line (e.g., red and blue dots on opposite sides).

Non-Linear SVM ‚Üí Used when data is more complex and cannot be separated by a straight line. In such cases, SVM uses a kernel trick to create a curved or more complex boundary.

***‚ú® Why SVM is useful:***

‚úÖ Finds a clear separation between classes

‚úÖ Works well for both linear and non-linear boundaries

‚úÖ Focuses only on important points (support vectors) instead of all data

‚ö†Ô∏è Might be slower for very large datasets


```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn import datasets
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC

# 1. Load a simple dataset (make_classification for beginners)
from sklearn.datasets import make_blobs
X, y = make_blobs(n_samples=50, centers=2, random_state=42, cluster_std=1.2)

# 2. Train SVM model
model = SVC(kernel='linear')
model.fit(X, y)

# 3. Plot decision boundary
plt.figure(figsize=(8,6))

# Scatter plot of points
plt.scatter(X[:,0], X[:,1], c=y, cmap='bwr', s=60, edgecolors='k')

# Plot decision boundary (line)
ax = plt.gca()
xlim = ax.get_xlim()
ylim = ax.get_ylim()

# Create grid
xx = np.linspace(xlim[0], xlim[1], 30)
yy = np.linspace(ylim[0], ylim[1], 30)
YY, XX = np.meshgrid(yy, xx)
xy = np.vstack([XX.ravel(), YY.ravel()]).T
Z = model.decision_function(xy).reshape(XX.shape)

# Draw decision boundary and margins
ax.contour(XX, YY, Z, colors='k',
           levels=[-1, 0, 1], alpha=0.7,
           linestyles=['--', '-', '--'])

# Highlight support vectors
ax.scatter(model.support_vectors_[:, 0],
           model.support_vectors_[:, 1],
           s=150, linewidth=1, facecolors='none', edgecolors='k')

plt.title("SVM with Linear Kernel (Beginner Example)")
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_circles
from sklearn.svm import SVC

# 1. Create circular dataset
X, y = make_circles(n_samples=100, factor=0.5, noise=0.05, random_state=42)

# 2. Train SVM with RBF kernel
model = SVC(kernel='rbf', C=1, gamma=0.5)
model.fit(X, y)

# 3. Plot decision boundary
plt.figure(figsize=(8,6))
plt.scatter(X[:,0], X[:,1], c=y, cmap='bwr', s=60, edgecolors='k')

# Create grid
ax = plt.gca()
xlim = ax.get_xlim()
ylim = ax.get_ylim()
xx = np.linspace(xlim[0], xlim[1], 200)
yy = np.linspace(ylim[0], ylim[1], 200)
YY, XX = np.meshgrid(yy, xx)
xy = np.vstack([XX.ravel(), YY.ravel()]).T
Z = model.decision_function(xy).reshape(XX.shape)

# Draw decision boundary
ax.contour(XX, YY, Z, colors='k',
           levels=[-1, 0, 1], alpha=0.7,
           linestyles=['--', '-', '--'])

# Highlight support vectors
ax.scatter(model.support_vectors_[:, 0],
           model.support_vectors_[:, 1],
           s=150, linewidth=1, facecolors='none', edgecolors='k')

plt.title("SVM with RBF Kernel (Non-Linear Example)")
plt.show()
```

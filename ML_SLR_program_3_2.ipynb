```python
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
import numpy as np

# Health data: Daily activity vs Blood sugar level
X = np.array([10, 20, 30, 40, 50]).reshape(-1, 1)  # Daily activity in minutes
y = np.array([190, 180, 160, 150, 140])           # Blood sugar level in mg/dL

# Create and train the linear regression model
model = LinearRegression()
model.fit(X, y)

# Predict for plotting the regression line
y_pred = model.predict(X)

# Plotting the scatter diagram and regression line
plt.figure(figsize=(8, 5))
plt.scatter(X, y, color='green', label='Patient Data')
plt.plot(X, y_pred, color='red', linewidth=2, label='Regression Line')
plt.title("Simple Linear Regression: Activity vs Blood Sugar Level")
plt.xlabel("Daily Physical Activity (minutes)")
plt.ylabel("Blood Sugar Level (mg/dL)")
plt.grid(True)
plt.legend()
plt.show()
```

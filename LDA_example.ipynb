```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis

# Dataset from image
X1 = np.array([[4,2],[2,4],[2,3],[3,6],[4,4]])   # Class 1
X2 = np.array([[9,10],[6,8],[9,5],[8,7],[10,8]]) # Class 2

# Combine into dataset
X = np.vstack((X1, X2))
y = np.array([0]*len(X1) + [1]*len(X2))  # Labels: 0 for 1, 1 for 2

# Apply LDA
lda = LinearDiscriminantAnalysis(n_components=1)
X_lda = lda.fit_transform(X, y)

# Plot original data in 2D
plt.figure(figsize=(12,5))

plt.subplot(1,2,1)
plt.scatter(X1[:,0], X1[:,1], c='blue', label='Class 1')
plt.scatter(X2[:,0], X2[:,1], c='red', label='Class 2')
plt.title("Original 2D Data")
plt.xlabel("x1")
plt.ylabel("x2")
plt.legend()

# Plot LDA 1D projection
plt.subplot(1,2,2)
plt.scatter(X_lda[y==0], np.zeros(len(X1)), c='blue', label='Class 1')
plt.scatter(X_lda[y==1], np.zeros(len(X2)), c='red', label='Class 2')
plt.axhline(0, color='black', linestyle='--')
plt.title("LDA Projection (1D)")
plt.xlabel("Projected axis")
plt.legend()

plt.tight_layout()
plt.show()
```

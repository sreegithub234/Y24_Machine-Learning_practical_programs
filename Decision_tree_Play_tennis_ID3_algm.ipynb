implement the Decision Tree (ID3 algorithm) on the classic Play Tennis dataset and also visualize the tree.

```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder
from sklearn.tree import DecisionTreeClassifier, plot_tree
import matplotlib.pyplot as plt
```

```python
# Step 1: Dataset
data = {
    'Outlook': ['Sunny','Sunny','Overcast','Rain','Rain','Rain','Overcast','Sunny',
                'Sunny','Rain','Sunny','Overcast','Overcast','Rain'],
    'Temp': ['Hot','Hot','Hot','Mild','Cool','Cool','Cool','Mild',
             'Cool','Mild','Mild','Mild','Hot','Mild'],
    'Humidity': ['High','High','High','High','Normal','Normal','Normal','High',
                 'Normal','Normal','Normal','High','Normal','High'],
    'Wind': ['Weak','Strong','Weak','Weak','Weak','Strong','Strong','Weak',
             'Weak','Weak','Strong','Strong','Weak','Strong'],
    'PlayTennis': ['No','No','Yes','Yes','Yes','No','Yes','No',
                   'Yes','Yes','Yes','Yes','Yes','No']
}

df = pd.DataFrame(data)
```

```python
# Step 2: Encode categorical features
le_dict = {}
for column in df.columns:
    le = LabelEncoder()
    df[column] = le.fit_transform(df[column])
    le_dict[column] = le
```

```python
# Step 3: Define features (X) and target (y)
X = df[['Outlook','Temp','Humidity','Wind']]
y = df['PlayTennis']
```

```python
# Step 4: Train Decision Tree using ID3 (entropy)
clf = DecisionTreeClassifier(criterion='entropy')
clf.fit(X, y)
```

```python
# Step 5: Plot the tree
plt.figure(figsize=(12,8))
plot_tree(
    clf,
    feature_names=['Outlook','Temp','Humidity','Wind'],
    class_names=le_dict['PlayTennis'].classes_,
    filled=True, rounded=True
)
plt.title("Decision Tree - ID3 (PlayTennis)")
plt.show()
```

Compute and Compare R-Square amd Adjusted R-Square values

```python
# Step 1: Import required libraries
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
```

```python
# Step 2: Define the dataset
# X = number of hours of TV watched (independent variable)
# y = test scores of students (dependent variable)
X = np.array([0, 1, 2, 3, 3, 5, 5, 5, 6, 7, 7, 10]).reshape(-1, 1)
y = np.array([96, 85, 82, 74, 95, 68, 76, 84, 58, 65, 75, 50])
```

```python
# Step 3: Fit the linear regression model
model = LinearRegression()
model.fit(X, y)
```

```python
# Step 4: Extract slope, intercept, predictions
intercept = model.intercept_      # β₀ (baseline value)
slope = model.coef_[0]           # β₁ (rate of change)
y_pred = model.predict(X)        # predicted values for y
r2 = r2_score(y, y_pred)         # coefficient of determination
```

```python
# Display regression equation
print(f"Regression equation: y = {intercept:.2f} + ({slope:.2f})x")
```

```python
# Step 4a: Compute Adjusted R²
n = len(y)          # number of observations
p = X.shape[1]      # number of predictors
adj_r2 = 1 - (1-r2)*(n-1)/(n-p-1)
```

```python
# Step 5: Display regression equation
print(f"Regression equation: y = {intercept:.2f} + ({slope:.2f})x")
```

```python
# Step 6: Print numerical results
print(f"Intercept (β₀): {intercept:.2f}")
print(f"Slope (β₁): {slope:.2f}")
print(f"R² Score: {r2:.3f}")
print(f"Adjusted R² Score: {adj_r2:.3f}")
```

```python
# Step 7: Interpretation of results
print("\nInterpretation:")
print(f"- Intercept (β₀ = {intercept:.2f}): Expected test score when TV hours = 0.")
print(f"- Slope (β₁ = {slope:.2f}): Each additional TV hour reduces the score by {abs(slope):.2f} points.")
print(f"- R² = {r2:.3f}: {r2*100:.1f}% of score variation is explained by TV hours.")
print(f"- Adjusted R² = {adj_r2:.3f}: Corrects R² for sample size and predictors. In simple regression, it is almost same as R².")
```

```python
# Step 8: Prediction for 9 hours of TV watched
pred_9 = model.predict(np.array([[9]]))[0]
print(f"\nPredicted test score for 9 hours of TV: {pred_9:.2f}")
```

**Interpretation of results with numbers**

If a student watches 0 hours of TV, their expected score ≈ 93.97.

Each extra 1 hour of TV reduces the score by about 4.07 points.

About 69.1% of score variation is explained by TV hours (R² = 0.691).

Adjusted R² = 0.660, slightly lower, meaning the model remains a fairly good fit considering the sample size.

A student watching 9 hours of TV is predicted to score only 57.36.
